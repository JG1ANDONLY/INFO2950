```python
import warnings
warnings.simplefilter(action='ignore')
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
sns.set_style('white')
%matplotlib inline
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
print(df.columns)
```

Because I want to build models separately for companies in the United States and those in foreign countries, I will create a new column named `USA` based on `location` in `df`. If the observation works in a technology company whose `location` is in the United States, the corresponding value of `USA` is 1, otherwise 0.

```python
df['USA'] = ''
for i in df.index:
    if df.loc[i, 'location'][-2].isupper():
        df.loc[i, 'USA'] = 1
    else:
        df.loc[i, 'USA'] = 0
```

```python
df.sample(5)
```

If not, I will use a multiple linear regression model for these variables.

```python
# initialize plotting alignments
matplotlib.style.use('default')
plt.figure(figsize = (15, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)
plt.style.use('ggplot')

# yearsofexperience
X = df_USA_for_linear['yearsofexperience'].values.reshape(-1, 1)
y = df_USA_for_linear['totalyearlycompensation'].values
model = LinearRegression().fit(X, y)
resp = model.predict(X)
# # r^2
r2 = model.score(X, y)
# # regression plot
ax1.plot(X, resp, color='orange', label='Regression Line')
ax1.scatter(X, y, edgecolor='k', facecolor='grey', alpha=0.5, label='true data')
ax1.set_xlabel('yearsofexperience')
ax1.set_ylabel('totalyearlycompensation')
ax1.legend(facecolor='white', fontsize=11)
ax1.set_title('yearsofexperience, $r^2= %.2f$' % r2, fontsize=18);

# yearsatcompany
X = df_USA_for_linear['yearsatcompany'].values.reshape(-1, 1)
y = df_USA_for_linear['totalyearlycompensation'].values
model = LinearRegression().fit(X, y)
resp = model.predict(X)
# # r^2
r2 = model.score(X, y)
# # regression plot
ax2.plot(X, resp, color='orange', label='Regression Line')
ax2.scatter(X, y, edgecolor='k', facecolor='grey', alpha=0.5, label='true data')
ax2.set_xlabel('yearsatcompany')
ax2.set_ylabel('totalyearlycompensation')
ax2.legend(facecolor='white', fontsize=11)
ax2.set_title('yearsatcompany, $r^2= %.2f$' % r2, fontsize=18);

# basesalary
X = df_USA_for_linear['basesalary'].values.reshape(-1, 1)
y = df_USA_for_linear['totalyearlycompensation'].values
model = LinearRegression().fit(X, y)
resp = model.predict(X)
# # r^2
r2 = model.score(X, y)
# # regression plot
ax3.plot(X, resp, color='orange', label='Regression Line')
ax3.scatter(X, y, edgecolor='k', facecolor='grey', alpha=0.5, label='true data')
ax3.set_xlabel('basesalary')
ax3.set_ylabel('totalyearlycompensation')
ax3.legend(facecolor='white', fontsize=11)
ax3.set_title('basesalary, $r^2= %.2f$' % r2, fontsize=18);

# bonus
X = df_USA_for_linear['bonus'].values.reshape(-1, 1)
y = df_USA_for_linear['totalyearlycompensation'].values
model = LinearRegression().fit(X, y)
resp = model.predict(X)
# # r^2
r2 = model.score(X, y)
# # regression plot
ax4.plot(X, resp, color='orange', label='Regression Line')
ax4.scatter(X, y, edgecolor='k', facecolor='grey', alpha=0.5, label='true data')
ax4.set_xlabel('bonus')
ax4.set_ylabel('totalyearlycompensation')
ax4.legend(facecolor='white')
ax4.set_title('bonus, $r^2= %.2f$' % r2, fontsize=18);
```

#### Multiple Linear Regression (MLR) for variables of interest

- United States

I will fit a multiple linear regression model using elements of `var_for_linear` as predictor variables and `totalyearlycompensation` as a response variable in `df_USA_for_linear`. Then, I will store the model in `USA_mult_model` and predicted `totalyearlycompensation` values in `USA_resp`, and make a new column in `df_USA_for_linear` named `pred`.

```python
USA_mult_model = LinearRegression().fit(df_USA_for_linear[var_for_linear], \
                                        df_USA_for_linear['totalyearlycompensation'])
USA_resp = USA_mult_model.predict(df_USA_for_linear[var_for_linear])
```

```python
df_USA_for_linear['pred'] = USA_resp
```

    - Coefficient & Intercept

I will make a dictionary named `USA_mult_coef_dict` containing coefficient for each predictor variable and the intercept of the fitted multiple linear regression model `USA_mult_model`, and create a pandas DataFrame called `df_USA_mult_coef` from `USA_mult_coef_dict` to understand coefficients and intercept of `USA_mult_model` better.

```python
USA_mult_coef_dict = {}
for i in range(len(var_for_linear)):
    USA_mult_coef_dict[var_for_linear[i]] = round(USA_mult_model.coef_[i], 2)
USA_mult_coef_dict['Intercept'] = round(USA_mult_model.intercept_, 2)
df_USA_mult_coef = pd.Series(USA_mult_coef_dict)
df_USA_mult_coef
```

    - Equation

$$\hat{\text{yearly salary}}= 1153.15 * \text{years of experience} - 816.96 * \text{years at company} + 1.67 * \text{base salary} + 1.4 * \text{bonus} - 62360.91$$

    - Visualization

I will first plot the slope (coefficient) for each predictor variable in `USA_mult_model`.

```python
b = list(USA_mult_model.coef_)
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_facecolor("white")
ax.plot(var_for_linear, b)
ax.set_ylabel("Slopes")
ax.set_title('Multiple Linear Regression slopes');
```

Next, I will plot the true data of `totalyearlycompensation` and predicted values of `totalyearlycompensation` and observe whether the predicted values fit the true data well.

```python
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_facecolor("white")
ax.plot(df_USA_for_linear['pred'], alpha=0.3, label = "predicted yearly salary", color='orange');
ax.plot(df_USA_for_linear['totalyearlycompensation'], alpha=0.5, label = 'true yearly salary', color='#1f77b4');
ax.legend();
```

I think there is one outlier in the data, on the rightmost, because I believe the true value of `totalyearlycompensation` of this observation is so high that it deviates enormously from the predicted value. Except for this outlier, predicted values of `totalyearlycompensation` are highly overlapping with their true data, which indicates that we successfully built the multiple linear regression model.

### Evaluation of significance

- the United States Multiple Linear regression model (`USA_mult_model`)

I will conduct hypothesis testings to test if the multiple linear regression model `USA_mult_model` is reliable.<br>

Hypotheses:<br>

1. H<sub>0</sub>: `yearsofexperience` does not contribute to `totalyearlycompensation` ($\beta$<sub>1</sub> = 0)
<br>H<sub>1</sub>: `yearsofexperience` contributes to `totalyearlycompensation` ($\beta$<sub>1</sub> $\not =$ 0)
<br><br>
2. H<sub>0</sub>: `yearsatcompany` does not contribute to `totalyearlycompensation` ($\beta$<sub>2</sub> = 0)
<br>H<sub>1</sub>: `yearsatcompany` contributes to `totalyearlycompensation` ($\beta$<sub>2</sub> $\not =$ 0)
<br><br>
3. H<sub>0</sub>: `basesalary` does not contribute to `totalyearlycompensation` ($\beta$<sub>3</sub> = 0)
<br>H<sub>1</sub>: `basesalary` contributes to `totalyearlycompensation` ($\beta$<sub>3</sub> $\not =$ 0)
<br><br>
4. H<sub>0</sub>: `bonus` does not contribute to `totalyearlycompensation` ($\beta$<sub>4</sub> = 0)
<br>H<sub>1</sub>: `bonus` contributes to `totalyearlycompensation` ($\beta$<sub>4</sub> $\not =$ 0)

I will use the Permutation Test to support my hypothesis testing.

```python
def permute(input_array):
    permuted = input_array.copy().values
    np.random.shuffle(permuted)
    return pd.Series(permuted)
```

```python
obs_m = USA_mult_model.coef_

permuted_beta_count = [0, 0, 0, 0]
for i in range(500):
    permuted_model = LinearRegression().fit(df_USA_for_linear[var_for_linear], \
                                            permute(df_USA_for_linear['totalyearlycompensation']))
    m = permuted_model.coef_
    accum = 0
    for i in range(4):
        if abs(m[accum]) > abs(obs_m[accum]):
            permuted_beta_count[accum] += 1
        accum += 1

print('p-values:')
print('yearsofexperience: ' + str(permuted_beta_count[0]/500))
print('yearsatcompany: ' + str(permuted_beta_count[1]/500))
print('basesalary: ' + str(permuted_beta_count[2]/500))
print('bonus: ' + str(permuted_beta_count[3]/500))
```

We can observe **p-values** of all predictor variables are extremely small (<0.05), so we are confident to reject H<sub>1</sub> of each hypothesis testing. That is to say, `yearsofexperience`, `yearsatcompany`, `basesalary`, and `bonus` all contribute to totalyearlycompensation. In this way, I have demonstrated the credibility of my multiple linear regression model `USA_mult_model`.
