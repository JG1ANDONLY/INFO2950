#### Multiple Linear Regression (MLR) for [`gender`, `edu_levels`, `race`]

In this part, I will perform multiple linear regression model building for `categorical_vars` as predictor variables with `totalyearlycompensation` as a response variable from data of `df_USA` and data of `df_foreign`, separately, to study the relationship between [`gender`, `edu_levels`, `race`] with `totalyearlycompensation` in US technology companies and foreign companies.

- United States

I will extract all columns of `df_USA` whose names are in `categorical_var` into a new pandas DataFrame named `cate_USA`.<br><br>
I will fit a multiple linear regression model using columns of `cate_USA` as predictor variables and `totalyearlycompensation` as a response variable. <br><br>
Then, I will store the model in `cate_USA_mult_model` and the predicted `totalyearlycompensation` values in a new column `pred` in `cate_USA`.

```python
cate_USA = df_USA[categorical_var]
cate_USA_mult_model = LinearRegression().fit(cate_USA, df_USA['totalyearlycompensation'])
cate_USA_resp = cate_USA_mult_model.predict(cate_USA)
cate_USA['pred'] = cate_USA_resp
```

    - Coefficient & Intercept

I will create a pandas DataFrame called `cate_USA_mult_coef` that contains coefficients and intercept of `cate_USA_mult_model`.

```python
coef = list(cate_USA_mult_model.coef_) + [cate_USA_mult_model.intercept_]
new_coef = []
for i in coef:
    new_coef.append(round(i, 2))
cate_USA_mult_coef = pd.DataFrame(new_coef, index=categorical_var+['intercept'], columns=['values'])
cate_USA_mult_coef
```

    - Equation

$$\hat{\text{yearly salary}}= - 13585.01 * \text{Female} + 13585.01 * \text{Male} - 61631.50 * \text{masters_degree} - 92592.86 * \text{bachelors_degree} - 128.42 * \text{doctorate_degree} - 73212.40 * \text{highschool} - 73643.92 * \text{some_college} - 36517.92 * \text{race_asian} - 29073.62 * \text{race_white} - 34676.05 * \text{race_two_or_more} - 62563.29 * \text{race_black} - 49520.54 * \text{race_hispanic} + 324046.30$$

    - Visualization

I will plot the true data of `totalyearlycompensation` and predicted values of `totalyearlycompensation` and observe whether the predicted values fit the true data well.

```python
fig = plt.figure(figsize=(15, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_facecolor("white")
ax.plot(cate_USA['pred'], alpha=0.3, label = "predicted yearly salary", color='orange');
ax.plot(df_USA['totalyearlycompensation'], alpha=0.5, label = 'true yearly salary', color='#1f77b4');
ax.legend();
```

**All predicted values of `totalyearlycompensation` deviate hugely from their true data, which indicates that this linear model may not be accurate.** <br><br>

    - Application

A person can predict his/her yearly salary by plugging their numbers into this equation.<br><br>
For example, a black female with a master degree who want to work at a US company can predict her yearly salaries this way:<br><br>
**yearly salary = - 13585.01 * 1 + 13585.01 * 0 - 61631.50 * 1 - 92592.86 * 0 - 128.42 * 0 - 73212.40 * 0 - 73643.92 * 0 - 36517.92 * 0 - 29073.62 * 0 - 34676.05 * 0 - 62563.29 * 1 - 49520.54 * 0 + 324046.30 = 186266.5**<br><br>
Her predicted yearly salary is 186266.5.
